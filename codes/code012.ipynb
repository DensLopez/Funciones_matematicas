# Perceptrn
Qu es?
Es la manera en que nosotros definimos a una neurona artificial.

Cmo funciona a nivel biologico y sencillo?
- Dentritas: Son mecanismos receptores de impulsos electricos 
  > Recuerda que todas las neuronas se comunican a traves de pequeos impulsos electricos, los cuales son procesados por el axn.
- Axn: Recoge todos los impulsos electricos que son captados tambin por el ncleo.
- Terminal del axn: Pasan por una serie de procesos a traves de los dems componentes y terminan por la terminal del axn.

Lo podemos ver como una funcin donde:
- Entran varios impulsos, sucede un proceso (Axn y demas componentes) y al final surge un mecanismo de salida, dependiendo de los procesos que surgieron en la parte anterior.
