# Cómo leer las matemáticas
## Símbolos generales

### Símbolos de igualdad o relación|
- \>\> Mucho mayor que 
- \<\< Mucho menor que
- ∞+ Número muy muy grande positivo
- ∞- Número muy muy grande negativo

### Símbolos de operaciones acumulativas
Son usadas para series o expansiones
- Σ Sigma - Sumatoria - Se puede ver como un ciclo for (tiene límite inferior y superior)
- ∏ Producto - Se puede modelar como un ciclo for (tiene límite inferior y superior)
